```python
# Definimos el estilo 
tabs_styles = {
    'height': '44px'
}

tab_style = {
    'borderBottom': '1px solid #555',  # L√≠nea inferior m√°s oscura
    'padding': '6px',
    'fontWeight': 'bold',
    'color': 'white',  # Texto en blanco
    'background-color': 'black'  # Fondo negro
}

tab_selected_style = {
    'borderTop': '1px solid #555',
    'borderBottom': '1px solid #555',
    'backgroundColor': 'black',  # Fondo seleccionado negro
    'color': 'white',
    'padding': '6px'
}


```

```python
# Step 1. Create the app
app = Dash(__name__)
# Step 2. Set the title
app.title = "Visualizaci√≥n KPIs"
# Step 3. Set the layout
app.layout = html.Div([
    html.H1('Visualizaci√≥n KPIs'),
    dcc.Tabs(id="tabs-example-graph", value='tab_1', children=[
        dcc.Tab(label='Visualizaci√≥n global de la situaci√≥n de las ventas',
                value='tab_1',
                style=tab_style,
                selected_style=tab_selected_style,
                children=html.Div([
                    html.H2(children='Conteos Generales'),
                    dcc.Dropdown(id='dropdown1',  
                                 options=[{'label': measures_tab_1a[key], 'value': key} for key in measures_tab_1a.keys()],
                                 value='store_type'),  
                    dcc.Graph(id='counts-graph'),
                    html.H2(children='Top 10 m√°s vendidos'), 
                    dcc.Graph(id='top10-product-graph'),
                    html.H2(children='Distribuci√≥n de las ventas por tiendas'), 
                    dcc.Graph(id='sales-distribution'),
                    html.H2(children='Top 10 tipos de tiendas con ventas en productos en promoci√≥n'),
                    dcc.Graph(id='top-10-stores-onpromotion'),
                    html.H2(children='An√°lisis de la estacionalidad de las ventas'),
                    html.H3(children='Media de ventas por semana'),
                    dcc.Graph(id='sales-mean-per-weekday'),
                    html.H2(children='Ventas semanales y mensuales por a√±o'),
                    html.H4(children='Seleccionar el a√±o: '),
                    dcc.Dropdown(id='dropdown3',  
                                 options=[{'label': year, 'value': year} for year in years_1c],
                                 value=2013), 
                    html.H3(children='Ventas mensuales'), 
                    dcc.Graph(id='sales-volume-per-week'),  
                    html.H3(children='Ventas semanales'), 
                    dcc.Graph(id='sales-volume-per-month')

                ], style={'width': '60%', 'height': '80vh', 'display': 'inline-block'})),
        
        dcc.Tab(label='Informaci√≥n por tienda',
                value='tab_2',
                style=tab_style,
                selected_style=tab_selected_style,
                children=html.Div([
                    html.H3(children='Seleccionar la tienda: '),
                    dcc.Dropdown(id='dropdown4',  
                                 options=[{'label': store, 'value': store} for store in df['store_nbr'].sort_values().unique()],
                                 value=1), dcc.Graph(id='sales-volume-per-year'), dcc.Graph(id='total-sales'), dcc.Graph(id='total-promotion-sales')
                ], style={'width': '60%', 'height': '80vh', 'display': 'inline-block'}),
        ),
        dcc.Tab(label='Informaci√≥n por estado',
                value='tab_3',
                style=tab_style,
                selected_style=tab_selected_style,
                children=html.Div([
                    html.H3(children='Seleccionar la el estado: '),
                    dcc.Dropdown(id='dropdown5',  
                                 options=[{'label': state, 'value': state} for state in df['state'].sort_values().unique()],
                                 value='Pichincha'), dcc.Graph(id='transactions-per-year'), dcc.Graph(id='top-sales-stores'), dcc.Graph(id='most-sold')
                ], style={'width': '60%', 'height': '80vh', 'display': 'inline-block'}),
        ),

     dcc.Tab(label='Visualizaciones Adicionales',
                value='tab_4',
                style=tab_style,
                selected_style=tab_selected_style,
                children=html.Div([html.H3(children='Series temporales de ventas por tipo de tienda: '),
                    html.H4(children='Seleccionar la tienda: '),
                    dcc.Dropdown(id='dropdown2',  
                                 options=[{'label': store, 'value': store} for store in stores_1b],
                                 value='A'), dcc.Graph(id='sales-distribution-per-type')
                ], style={'width': '60%', 'height': '80vh', 'display': 'inline-block'}),
        ),

    ]),
    
    
    html.Div(id='tabs-content-example-graph')
])

# Step 3. Set the callback functions
# Callback del primer dropdown 
@app.callback(
    [Output(component_id='counts-graph', component_property='figure'),
     Output(component_id='top10-product-graph', component_property='figure'),Output('sales-distribution', 'figure'),Output('top-10-stores-onpromotion', 'figure'),Output('sales-mean-per-weekday', 'figure')],
    Input(component_id='dropdown1', component_property='value')
)
def update_graph(selected_measure):
    if selected_measure != 'family':
        counts = df[selected_measure].value_counts().sort_index().dropna()
        values = counts.values
    else:
        counts = df[['family','sales']].groupby('family').sum() # Si el atributo seleccionado es family hacemos un conteo de las ventas 
        values = counts.values.tolist()
        values = [float(item[0]) for item in values ]
    index = list(counts.index)
    if type(index[0]) != str:
        index = list(map(str, index))
    data = {'index': index, 'values': values}
    counts_df = pd.DataFrame(data)
    hist_fig = px.histogram(counts_df,
                            x='index',
                            y='values',
                            title=f'Histograma de {selected_measure}',
                            labels={'index': selected_measure, 'values': 'counts'},
                            barmode='group'
                            ,
                            color_discrete_sequence=color_scale
                           )
    if selected_measure != 'family' and selected_measure != 'month':
        hist_fig.update_layout(annotations=[
            dict(
                x=x_val,
                y=y_val,
                text=str(y_val),
                showarrow=True,
                arrowhead=4,
                ax=0,
                ay=-40
            ) for x_val, y_val in zip(index, values)
        ])
    most_sold_df = df[['family','sales']].groupby('family').sum().sort_values(by='sales', ascending=False)
    index = list(most_sold_df.index)
    values = most_sold_df.values.tolist()
    values = [float(item[0]) for item in values ]
    data = {'index': index[:10], 'values': values[:10]}
    most_sold_df = pd.DataFrame(data)

    most_sold = px.histogram(most_sold_df,
                             x='index',
                             y='values',
                             title=f'Histograma top 10 productos vendidos',
                             labels={'index': 'Tipo de producto', 'values': 'conteo de ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )
    most_sold.update_layout(annotations=[
            dict(
                x=x_val,
                y=y_val,
                text=str(int(y_val)),
                showarrow=True,
                arrowhead=4,
                ax=0,
                ay=-40
            ) for x_val, y_val in zip(index[:10], values[:10])
        ])
    df_sales_per_store = df.groupby('store_nbr')['sales'].sum()
    index = list(df_sales_per_store.index)
    index = [str(i) for i in index]
    values = df_sales_per_store.values.tolist()
    data = {'index': index, 'values': values}
    df_sales_per_store = pd.DataFrame(data)
    sales_per_store = px.histogram(df_sales_per_store,
                             x='index',
                             y='values',
                             title=f'Histograma de ventas por tienda',
                             labels={'index': 'Numero de tiendas', 'values': 'conteo de ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )


    promotion_df = df[df['onpromotion'] != 0].groupby('store_nbr')['onpromotion'].sum().sort_values(ascending=False)
    index = list(promotion_df.index)
    index = [str(i) for i in index]
    values = promotion_df.values.tolist()
    data = {'index': index[:10], 'values': values[:10]}
    promotion_df = pd.DataFrame(data)
    promotion = px.histogram(promotion_df,
                             x='index',
                             y='values',
                             title=f'Histograma top tiendas con ventas en productos en promoci√≥n',
                             labels={'index': 'Tienda', 'values': 'conteo de ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )
    sales_mean_per_weekday_df = df.groupby('day_of_week')['sales'].mean().sort_values(ascending=False)
    index = list(sales_mean_per_weekday_df.index)
    values = sales_mean_per_weekday_df.values.tolist()
    data = {'index': index, 'values': values}
    sales_mean_per_weekday_df = pd.DataFrame(data)
    sales_mean_per_weekday = px.histogram(sales_mean_per_weekday_df,
                             x='index',
                             y='values',
                             title=f'Media de ventas por d√≠a de la semana',
                             labels={'index': 'Dia de la semana', 'values': 'media de ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )

    return [hist_fig, most_sold,sales_per_store,promotion,sales_mean_per_weekday]



# Callback del tercer dropdown
@app.callback(
    [Output('sales-volume-per-week', 'figure'),Output('sales-volume-per-month', 'figure')],
    Input('dropdown3', 'value') 
    )
def update_graph(year):
    df_year = df[df['year'] == year]

    sales_per_week_df = df_year.groupby('week')['sales'].sum()
    index = list(sales_per_week_df.index)
    values = sales_per_week_df.values.tolist()
    data = {'index': index, 'values': values}
    sales_per_week_df = pd.DataFrame(data)
    sales_per_week = px.line(sales_per_week_df,
                      x='index',
                      y='values',
                      title=f'Volumen de ventas semanales de {year}',
                            color_discrete_sequence=color_scale
                           )
    sales_per_month_df = df_year.groupby('month')['sales'].sum()
    index = list(sales_per_month_df.index)
    values = sales_per_month_df.values.tolist()
    data = {'index': index, 'values': values}
    sales_per_month_df = pd.DataFrame(data)
    sales_per_month = px.line(sales_per_month_df,
                      x='index',
                      y='values',
                      title=f'Volumen de ventas mensuales de {year}',
                            color_discrete_sequence=color_scale
                           )


    return [sales_per_week,sales_per_month]

@app.callback(
    [Output('sales-volume-per-year', 'figure'),Output('total-sales', 'figure'),Output('total-promotion-sales', 'figure')],
    Input('dropdown4', 'value') 
    )
def update_graph(store_number):
    df_store = df[df['store_nbr'] == store_number]

    df_sales_year = df_store.groupby('year')['sales'].sum().sort_index()
    index = list(df_sales_year.index)
    index =  [str(i) for i in index]
    values = df_sales_year.values.tolist()
    data = {'index': index, 'values': values}
    df_sales_year = pd.DataFrame(data)
    sales_year = px.histogram(df_sales_year,
                             x='index',
                             y='values',
                             title=f'Ventas por a√±o de la tienda {store_number}',
                             labels={'index': 'Dia de la semana', 'values': 'media de ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )
    df_total_products = df_store.groupby('family')['sales'].sum()
    index = list(df_total_products.index)
    values = df_total_products.values.tolist()
    data = {'index': index, 'values': values}
    df_total_products = pd.DataFrame(data)
    total_products = px.histogram(df_total_products,
                             x='index',
                             y='values',
                             title=f'Productos vendidos en la tienda {store_number}',
                             labels={'index': 'tipo de producto', 'values': 'ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )
    
    df_total_promotion_products = df_store[df_store['onpromotion'] != 0 ].groupby('family')['sales'].sum()
    index = list(df_total_promotion_products.index)
    values = df_total_promotion_products.values.tolist()
    data = {'index': index, 'values': values}
    df_total_promotion_products = pd.DataFrame(data)
    total_promotion_products = px.histogram(df_total_products,
                             x='index',
                             y='values',
                             title=f'Productos en promoci√≥n vendidos en la tienda {store_number}',
                             labels={'index': 'tipo de producto', 'values': 'ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )
    
    return [sales_year,total_products,total_promotion_products]



dcc.Graph(id='transactions-per-year'), dcc.Graph(id='top-sales-stores'), dcc.Graph(id='most-sold')
@app.callback(
    [Output('transactions-per-year', 'figure'),Output('top-sales-stores', 'figure'),Output('most-sold', 'figure')],
    Input('dropdown5', 'value') 
    )
def update_graph(state):
    df_state = df[df['state'] == state]
    df_total_transactions = df_state.groupby('year')['transactions'].sum().sort_index()
    index = list(df_total_transactions.index)
    index =  [str(i) for i in index]
    values = df_total_transactions.values.tolist()
    data = {'index': index, 'values': values}
    df_total_transactions = pd.DataFrame(data)
    transactions = px.histogram(df_total_transactions,
                            x='index',
                            y='values',
                            title=f'Transacciones por a√±o en el estado de {state}',
                            labels={'index': 'A√±o', 'values': 'Transacciones'},
                            barmode='group',
                            color_discrete_sequence=color_scale
                           )
    df_most_sales = df_state.groupby('store_nbr')['sales'].sum().sort_values(ascending=False)
    index = list(df_most_sales.index)
    index =  [str(i) for i in index]
    values = df_most_sales.values.tolist()
    data = {'index': index, 'values': values}
    df_most_sales = pd.DataFrame(data)
    most_sales = px.histogram(df_most_sales,
                             x='index',
                             y='values',
                             title=f'Ventas por tienda en el estado de {state}',
                             labels={'index': 'Tienda', 'values': 'ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )
    
    df_most_sold_product = df_state.groupby(['store_nbr','family'])['sales'].sum()
    store_numbers = np.sort(df_state['store_nbr'].unique())
    most_sold_products = []
    for store_number in store_numbers:
        serie = df_most_sold_product.xs(key=store_number)
        valor_buscado = serie.values.max()
        most_sold_products.append(serie.loc[serie == valor_buscado].index[0])


    # Add annotations for most_sold_products
    for i, product in enumerate(most_sold_products):
        most_sales.add_annotation(
            x=i,
            y=df_most_sales.loc[df_most_sales['index'] == index[i], 'values'].values[0],
            text=product,
            showarrow=True,
            arrowhead=4,
            ax=0,
            ay=-40
        )   
    
    df_most_sold = df_state.groupby('family')['sales'].sum().sort_values(ascending=False)
    index = list(df_most_sold.index)
    index =  [str(i) for i in index]
    values = df_most_sold.values.tolist()
    data = {'index': index, 'values': values}
    df_most_sold = pd.DataFrame(data)
    most_sold = px.histogram(df_most_sold,
                             x='index',
                             y='values',
                             title=f'Productos m√°s vendidos en el estado de {state}',
                             labels={'index': 'Producto', 'values': 'ventas'},
                             barmode='group'
                             ,
                            color_discrete_sequence=color_scale
                           )
    return [transactions,most_sales,most_sold]



# Callback del segundo dropdown 
@app.callback(
    [Output('sales-distribution-per-type', 'figure')],
    Input('dropdown2', 'value')
)
def update_sales_distribution(selected_store):
    df_company = df[df['store_type'] == selected_store]
    # Convert 'date' column to datetime format if it's not already
    df_company['date'] = pd.to_datetime(df_company['date'])
    df_company.set_index('date', inplace=True)
    df_copy = df_company.copy()
    df_monthly = df_copy.resample('M').sum().reset_index()
    dist = px.line(df_monthly,
                x='date',
                y='sales',
                title=f'Serie temporal de ventas de la tiendas tipo {selected_store}',
                            color_discrete_sequence=color_scale
                           )
    dist.update_xaxes(
        dtick="M1",
        tickformat="%b\n%Y",
        ticklabelmode="period"
    )
    return [dist] # Devuelve la figura dentro de una lista (si no da error)



    
# Step 4. Function to run the app in the external mode
app.run_server(jupyter_mode="external")
```
