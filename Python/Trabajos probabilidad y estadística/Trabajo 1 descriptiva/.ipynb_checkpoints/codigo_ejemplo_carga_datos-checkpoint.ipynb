## Cargamos los datos con los que se trabaja

```python
nom_fi_datos = 'G02B_DATOS_DEM_M4.csv'
```

```python
df_orig = pd.read_csv(nom_fi_datos)

# parseamos la fecha (cadena) para que sea un datetime con formato dd/mm/yyyy 
df_orig.FECHA = pd.to_datetime(df_orig.FECHA, format='%d/%m/%Y') 
s=df_orig.loc[:,'FECHA']
df_orig['DATE'] =  s.dt.date
del df_orig['FECHA']

# añadimos el año-semana para luego poder trabajar cómodamente con el dataframe
ywd=s.dt.isocalendar()
yw = 1000*ywd['year'] + ywd['week'] # por ejemplo 2022032 es la semana 32 del año 2022
df_orig['ANNO-SEMANA'] = yw

print('Tamaño de df_orig con los datos cargados:', df_orig.shape)
print('Dataframe con todos los datos leidos (filas: días, columnas: variables):')
print(df_orig)
```

### 1. Visión general de las series

```python
# auxiliares útiles para la representación
fechas = df_orig.DATE.tolist()
```

```python
fig = plt.figure(1, figsize=(15,6)) # permite indicar el nº de la figura y las dimensiones (ancho y alto)
ax1 = plt.subplot(2,1,1)
plt.plot(fechas, df_orig.CR00, label='CR00')
plt.plot(fechas, df_orig.CR12, label='CR12')
plt.plot(fechas, df_orig.CR20, label='CR20')
plt.title(nom_fi_datos)
plt.legend()

plt.subplot(2,1,2,sharex= ax1)
plt.plot(fechas, df_orig.TMAX, label='TMAX')
plt.plot(fechas, df_orig.TMIN, label='TMIN')
plt.tight_layout()
plt.legend()

plt.show()

fig.savefig('figura1.pdf')
```

12 y 20:
- ¿Cómo es la distribución de los coeficientes de reparto en las horas indicadas?
- ¿La distribución cambia con el día de la semana? ¿Y con el mes? ¿Y con la
festividad?
- ¿Cómo se relacionan los coeficientes de la hora 12 respecto a la temperatura
máxima? ¿Y respecto a la temperatura mínima? ¿La respuesta observada
depende del tipo de día?
- ¿Cómo se relacionan los coeficientes de la hora 20 respecto a la temperatura
máxima? ¿La respuesta observada depende de la época del año? ¿El coeficiente
de correlación lineal entre los coeficientes de la hora 20 y la temperatura cambia
significativamente según si en la hora 20 hay luz natural o no?

```python
CR12 = df_orig.CR12
CR20 = df_orig.CR20

me_12 = np.mean(CR12)    # media
md_12 = np.median(CR12)  # mediana
print('media(CR12):', me_12)
print('mediana(CR12):', md_12)

me_20 = np.mean(CR20)    # media
md_20 = np.median(CR20)  # mediana
print('media(CR20):', me_20)
print('mediana(CR20):', md_20)
print("La media y la mediana se parecen bastante, podríamos sacar de ahí que no parece que haya muchos atípicos")
```

```python
# mínimo y máximo
CR12_min = min(CR12)
CR12_max = max(CR12)
print('mínimo de CR12:', CR12_min, ', máximo de CR12:', CR12_max)

# mínimo y máximo
CR20_min = min(CR20)
CR20_max = max(CR20)
print('mínimo de CR20:', CR20_min, ', máximo de CR20:', CR20_max)
print("El mínimo de las 20 es más bajo que el de las 12, tiene sentido ya que por la noche suele hacer más frío")
```

```python
fig = plt.figure(1, figsize=(15,6)) # permite indicar el nº de la figura y las dimensiones (ancho y alto)
ax1 = plt.subplot(2,1,1)
plt.plot(fechas, df_orig.CR12, label='CR12')
plt.plot(fechas, df_orig.CR20, label='CR20')
plt.title("Evolución temporal de CR12 y CR20")
plt.legend()
```

```python
nbins = 250 # número de intervalos, ajustar para buena visualización

fig = plt.figure(1, figsize=(8, 4)) # permite indicar el nº de la figura y las dimensiones (ancho y alto)
sns.histplot(data=CR12, bins=nbins)
# sns.rugplot(x, color='r')                            # opcional
plt.title(f'Histograma ({nbins} intervalos)')
plt.xlabel('CR12')
plt.ylabel('Frecuencia absoluta')
plt.show()

nbins = 250 # número de intervalos, ajustar para buena visualización

fig = plt.figure(1, figsize=(8, 4)) # permite indicar el nº de la figura y las dimensiones (ancho y alto)
sns.histplot(data=CR20, bins=nbins)                            # opcional
plt.title(f'Histograma ({nbins} intervalos)')
plt.xlabel('CR20')
plt.ylabel('Frecuencia absoluta')
plt.show()
```

```python
# creamos dataframe para facilitar los cálculos y gráficos 
df =  pd.DataFrame(list(zip(CR12, CR20)), columns=['CR12', 'CR20'])
#print(df)

# matriz de covarianzas creada directamente con el dataframe (df)
mat_cov = df.cov()                
print('Matriz covarianzas:')
print(np.round(mat_cov, 2)) # Redondeando a solo dos decimales para verlo mejor

# matriz de correlaciones creada directamente con el dataframe (df)
mat_corr = df.corr()                
print('Matriz correlaciones:')
print(np.round(mat_corr, 2)) # Redondeando a solo dos decimales para verlo mejor

# Muestra la matriz de correlaciones visualmente
fig = plt.figure(1,figsize=(5, 5))
sns.heatmap(mat_corr, annot=True) # si no se ven los textos con demasiadas variables, poner annot=False
plt.title('Matriz de correlaciones')
plt.show()

print("Observamos que están poco correlacionadas, tiene sentido ya que son diferentes")
```

```python
nbins = 250 # número de intervalos histograma, ajustar para buena visualización

# Tres ejes en vertical para sincronizar la variable x
fig = plt.figure(1, figsize=(8, 6)) # permite indicar el nº de la figura y las dimensiones (ancho y alto)

# histograma
ax1= plt.subplot(3,1,1)
sns.histplot(data=CR12, bins=nbins)
plt.title('Distribución de CR12')
# cajas
ax2=plt.subplot(3,1,2, sharex=ax1)
sns.boxplot(data=CR12, orient='horizontal')
# violin
ax3=plt.subplot(3,1,3, sharex=ax1)
sns.violinplot(data=CR12, orient='horizontal')
sns.rugplot(data=CR12, color='r')
plt.xlabel('x')
plt.show()

nbins = 250 # número de intervalos histograma, ajustar para buena visualización

# Tres ejes en vertical para sincronizar la variable x
fig = plt.figure(1, figsize=(8, 6)) # permite indicar el nº de la figura y las dimensiones (ancho y alto)

# histograma
ax1= plt.subplot(3,1,1)
sns.histplot(data=CR20, bins=nbins)
plt.title('Distribución de CR20')
# cajas
ax2=plt.subplot(3,1,2, sharex=ax1)
sns.boxplot(data=CR20, orient='horizontal')
# violin
ax3=plt.subplot(3,1,3, sharex=ax1)
sns.violinplot(data=CR20, orient='horizontal')
sns.rugplot(data=CR20, color='r')
plt.xlabel('x')
plt.show()

print("La demanda de las 8 es mayor que la de las 12, esto tiene sentido ya que a las 8 hay más gente en sus casas que a las 12")
```
