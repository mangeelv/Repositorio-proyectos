```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

import funciones_auxiliares as fa
```

1. Exploratory Data Analysis (EDA) 

```python
df = pd.read_csv('FICO_Dataset.csv',delimiter=';')
```

```python
print(df.head())
```

**Gráficos de las variables + Medidas estadísticas básicas**

```python
variables = df.columns
for variable in variables:
    x = df[variable]
    nbins = 30
    # histograma
    ax1= plt.subplot(3,1,1)
    sns.histplot(data=x, bins=nbins)
    plt.title(f'Distribución de {variable}')
    # cajas
    ax2=plt.subplot(3,1,2, sharex=ax1)
    sns.boxplot(data=x, orient='horizontal')
    # violin
    ax3=plt.subplot(3,1,3, sharex=ax1)
    sns.violinplot(data=x, orient='horizontal')
    sns.rugplot(data=x, color='r')
    plt.xlabel('x')
    plt.show()

    me = np.mean(x)    # media
    md = np.median(x)  # mediana
    mo = stats.mode(x,axis=None, keepdims=True) # moda
    mo = mo.mode[0]
    print('Medidad estadísticas de:', variable)
    print(f'media({variable}):', me)
    print(f'mediana({variable}):', md)
    print(f'moda({variable}):', mo)
    rg = max(x)-min(x)  # rango
    iqr = stats.iqr(x) # rango intercuartílico
    va = np.var(x) # varianza
    st = np.std(x) # desviación típica
    cvar  = stats.variation(x) # coeficiente de variación
    print('variable:', variable)
    print(f'Rango({variable}):', rg)
    print(f'Rango intercuartílico IQR({variable}):', iqr)
    print(f'Varianza({variable}):', va)
    print(f'Desviación típica({variable}):', st)
    print(f'Coeficiente de variación({variable}):', cvar)
```

**Matriz de covarianzas y de correlaciones**

```python

# matriz de covarianzas creada directamente con el dataframe (df)
mat_cov = df.cov()                
print('Matriz covarianzas:')
print(np.round(mat_cov, 2)) # Redondeando a solo dos decimales para verlo mejor

# matriz de correlaciones creada directamente con el dataframe (df)
mat_corr = df.corr()                
print('Matriz correlaciones:')
print(np.round(mat_corr, 2)) # Redondeando a solo dos decimales para verlo mejor

# Muestra la matriz de correlaciones visualmente
fig = plt.figure(1,figsize=(15, 8))
sns.heatmap(mat_corr, annot=True) # si no se ven los textos con demasiadas variables, poner annot=False
plt.title('Matriz de correlaciones')
plt.show()
```

Eliminacion de valores nulos: Se porcede con la eliminación de las filas cuyo output es nulo ya que no aportan nada a la hora de crear modelos. Hay bastantes pero no se puede hacer nada. 

```python
df = df.dropna(subset=['RiskPerformance'])
print(df.info())
```

Eliminamos también filas con valores nulos en las otras columnas

```python
df = df.dropna()
print(df.info())
```

Valores especiales (-7,-8,-9)

```python
# Encontrar todas las filas que contienen -7, -8 o -9 en alguna de sus columnas
rows_with_values = df[(df == -7).any(axis=1) | (df == -8).any(axis=1) | (df == -9).any(axis=1)]

# Imprimir las filas encontradas
print(rows_with_values)
```

Estos valores indican que bien no hay registros del historial crediticio de la persona, bien no hay transacciones válidas o investigaciones de crédito o bien no se cumplen ciertas condiciones. Todos ellos denotan ausencia de información por lo que se procede a eliminarlos. 

```python
df = df.drop(index=rows_with_values.index)
print(df.info())
```

Tipificación de variables numéricas: Conviene que estén en el mismo rango

```python
column_no_tipify = 'RiskPerformance'
means = df.drop(columns=column_no_tipify).mean()
std_devs = df.drop(columns=column_no_tipify).std()
# Tipificar cada columna excepto la columna a no tipificar
df_standardized = df.drop(columns=column_no_tipify)
for column in df_standardized.columns:
    df_standardized[column] = (df_standardized[column] - means[column]) / std_devs[column]
df_standardized[column_no_tipify] = df[column_no_tipify]

df = df_standardized
```

Buscamos correlaciones significativas

```python
columnas_correlacionadas = set()
for i in range(len(mat_corr.columns)):
    for j in range(i):
        if abs(mat_corr.iloc[i, j]) >= 0.6: # Absoluto de la correlación para considerar tanto positivas como negativas
            colname1 = mat_corr.columns[i]
            colname2 = mat_corr.index[j]
            columnas_correlacionadas.add((colname1, colname2, mat_corr.iloc[i, j]))
print("Pares de columnas correlacionadas con correlación mayor que 0.6:")
for par in columnas_correlacionadas:
    print(par)
```
